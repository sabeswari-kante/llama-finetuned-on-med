```python

# 2. Merge Them
# We multiply patch_data * 5 to make sure the model REALLY learns these corrections
final_data = main_data + (patch_data * 5)

print(f"âœ… Merged Size: {len(final_data)} cases")

# 3. Save the Master File
with open("FINAL_MASTER_DATASET.json", "w") as f:
    json.dump(final_data, f, indent=2)

print("ðŸŽ‰ Created 'FINAL_MASTER_DATASET.json'. You are ready for the final run.")
```
