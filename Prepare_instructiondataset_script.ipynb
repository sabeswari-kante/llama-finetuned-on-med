```python
def generate_dataset(n_standard=100, n_adversarial=50):
    dataset = []
    
    # A. Generate Standard Cases
    print(f"Generating {n_standard * len(CASE_TEMPLATES)} Standard Cases...")
    for disease, t in CASE_TEMPLATES.items():
        for _ in range(n_standard):
            # Fill variables
            vars_dict = t["variables"].copy()
            # Resolve lists to single values
            context = {k: random.choice(v) for k, v in vars_dict.items()}
            
            # Helper to replace placeholders
            def fill(text):
                for k, v in context.items():
                    text = text.replace(f"{{{k}}}", str(v))
                return text

            age = random.randint(t["ages"][0][0], t["ages"][0][1])
            gender = random.choice(t["genders"])
            
            input_text = f"Patient: {age}y {gender}\nComplaint: {random.choice(t['chief_complaints'])}\nHistory: {fill(random.choice(t['histories']))}\nExam: {fill(random.choice(t['exam_findings']))}"
            
            reasoning = fill(t["reasoning"])
            diffs = "\n".join([f"{i+1}. {d[0]} - {d[1]}" for i, d in enumerate(t["differentials"])])
            invs = "\n".join([f"{i+1}. {d[0]} - {d[1]}" for i, d in enumerate(t["investigations"])])
            plan = "\n".join([f"{i+1}. {d[0]} - {d[1]}" for i, d in enumerate(t["referrals"])])
            
            output_text = f"DIAGNOSIS: {disease}\n\nREASONING: {reasoning}\n\nDIFFERENTIALS:\n{diffs}\n\nINVESTIGATIONS:\n{invs}\n\nPLAN/REFERRAL:\n{plan}"
            
            dataset.append({
                "instruction": "Act as a Medical Consultant. Diagnose and Plan.",
                "input": input_text,
                "output": output_text
            })

    # B. Generate Adversarial Cases
    print(f"Generating {n_adversarial * len(ADVERSARIAL_TEMPLATES)} Adversarial Cases...")
    for _ in range(n_adversarial):
        for t in ADVERSARIAL_TEMPLATES:
            output_text = f"DIAGNOSIS: {t['diagnosis']}\n\nREASONING: {t['reasoning']}\n\nPLAN/REFERRAL:\n{t['plan']}"
            dataset.append({
                "instruction": "Act as a Medical Consultant. Diagnose and Plan.",
                "input": t['input'],
                "output": output_text
            })
            
    random.shuffle(dataset)
    return dataset
```

```python
if __name__ == "__main__":
    full_data = generate_dataset(n_standard=100, n_adversarial=100)
    
    filename = "medical_instruction_dataset.json"
    with open(filename, 'w') as f:
        json.dump(full_data, f, indent=2)
        
    print(f"âœ… SUCCESS! Created {len(full_data)} cases in '{filename}'")
    print("Sample Reasoning:\n" + full_data[0]['output'].split('REASONING: ')[1].split('\n')[0])
```
