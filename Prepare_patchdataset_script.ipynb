```python
import json
import random
```

```python
PATCH_TEMPLATES = {
    # 1. CHICKENPOX (Targeting the "Scabies" hallucination)
    "Chickenpox (Varicella)": {
        "ages": [(5, 12)], "genders": ["male", "female"],
        "chief_complaints": ["itchy rash with blisters", "fever and spots all over"],
        "histories": ["Low grade fever for 2 days followed by a rash. Rash started on the trunk and spread to the face and limbs. Intense itching."],
        "exam_findings": ["Polymorphic rash (papules, vesicles, and crusts present simultaneously). Classic 'dew drop on a rose petal' appearance (clear fluid vesicle on red base). No burrows seen."],
        "reasoning": "The description of 'dew drop on a rose petal' vesicles is pathognomonic (specific) for Chickenpox. The polymorphic nature of the rash (different stages) confirms this. Scabies would present with burrows in web spaces, not clear vesicles on the trunk.",
        "differentials": [("Scabies", "Burrows in finger webs; intense night itch"), ("Impetigo", "Honey-colored crusts"), ("Measles", "Koplik spots; rash is maculopapular, not vesicular")],
        "investigations": [("Clinical Diagnosis", "Usually sufficient"), ("Tzanck Smear", "Multinucleated giant cells"), ("PCR", "Confirm VZV DNA")],
        "referrals": [("Home Isolation", "Contagious until crusted")],
        "variables": {}
    },
    
    # 2. DENGUE FEVER (Targeting the "Malaria" bias)
    "Dengue Fever": {
        "ages": [(15, 40)], "genders": ["male", "female"],
        "chief_complaints": ["severe eye pain and bone pain", "rash and high fever"],
        "histories": ["High fever of sudden onset. Severe retro-orbital (behind eye) pain. intense muscle and joint pain ('break-bone fever')."],
        "exam_findings": ["Temp 39C. Generalized maculopapular rash (islands of white in a sea of red). Tourniquet test positive (petechiae). Spleen NOT palpable."],
        "reasoning": "Retro-orbital pain + Rash + Positive Tourniquet test is the classic triad for Dengue. Crucially, Malaria rarely causes a rash and usually presents with splenomegaly, which is absent here. The 'break-bone' pain distinguishes it from Typhoid.",
        "differentials": [("Malaria", "Splenomegaly present; No rash"), ("Chikungunya", "Joint pain is more severe/chronic"), ("Typhoid", "Abdominal pain/Constipation")],
        "investigations": [("Dengue NS1 Antigen", "Diagnostic in early phase"), ("FBC", "Check for Thrombocytopenia (Low platelets)"), ("Malaria RDT", "To exclude co-infection")],
        "referrals": [("Medical Ward", "Fluid management")],
        "variables": {}
    },

    # 3. SCABIES (To teach what it ACTUALLY looks like)
    "Scabies": {
        "ages": [(5, 80)], "genders": ["male", "female"],
        "chief_complaints": ["severe itching especially at night"],
        "histories": ["Intense pruritus (itching) that worsens at night. Other family members also itching."],
        "exam_findings": ["Excoriations (scratch marks). Serpiginous burrows visible in the web spaces of fingers and flexor surfaces of wrists. No fluid-filled vesicles on trunk."],
        "reasoning": "The nocturnal dominance of itching and the presence of 'burrows' in finger webs are specific for Scabies. The absence of fever and distinct vesicles distinguishes this from Chickenpox.",
        "differentials": [("Chickenpox", "Febrile; vesicular rash"), ("Eczema", "History of atopy; no burrows"), ("Bed Bugs", "Linear bite marks")],
        "investigations": [("Skin Scraping", "Microscopy to see mites/eggs")],
        "referrals": [("GP", "Permethrin cream for whole family")],
        "variables": {}
    }
}

```

```python
def generate_patch(n=50):
    dataset = []
    print(f"ðŸ©¹ Generating {n * len(PATCH_TEMPLATES)} Patch Cases...")
    for disease, t in PATCH_TEMPLATES.items():
        for _ in range(n):
            # Simple filler
            age = random.randint(t["ages"][0][0], t["ages"][0][1])
            gender = random.choice(t["genders"])
            input_text = f"Patient: {age}y {gender}\nComplaint: {random.choice(t['chief_complaints'])}\nHistory: {t['histories'][0]}\nExam: {t['exam_findings'][0]}"
            
            output_text = f"DIAGNOSIS: {disease}\n\nREASONING: {t['reasoning']}\n\nDIFFERENTIALS:\n" + "\n".join([f"{i+1}. {d[0]} - {d[1]}" for i, d in enumerate(t['differentials'])]) + f"\n\nPLAN/REFERRAL:\n{t['referrals'][0][0]} - {t['referrals'][0][1]}"
            
            dataset.append({"instruction": "Act as a Medical Consultant. Diagnose and Plan.", "input": input_text, "output": output_text})
            
    return dataset
```

```python

patch_data = generate_patch(50)
with open("medical_patch_dataset.json", "w") as f:
    json.dump(patch_data, f, indent=2)
print("âœ… Patch Data Created!")
```
